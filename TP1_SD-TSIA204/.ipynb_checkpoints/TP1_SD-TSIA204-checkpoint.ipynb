### Question 3.

To perform linear regression for **Investment** on **GNP**, we will calculate the intercept, slope, their standard deviations, and the determination coefficient $R^2$ using **elementary arithmetic** (as libraries for regression are not allowed).

***

#### (a). Linear Regression Formula Recap
Let the log-transformed GNP be $x = \log(\text{GNP}), $and the log-transformed Investment be $y = \log(\text{Investment})$. The regression line is:
$$
\hat{y} = \hat{\beta_0} + \hat{\beta_1} x
$$
where:
- $ \hat{\beta_1} = \dfrac{\text{Cov}(x, y)}{\text{Var}(x)} = \dfrac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sum{(x_i - \bar{x})^2}} $
- $ \hat{\beta_0} = \bar{y} - \hat{\beta_1} \bar{x} $

***

#### (b). Standard Deviations of Intercept and Slope
The standard deviations are calculated as:
$$
\sqrt{\mathbb{V}(\hat{\beta_1})} = \sqrt{\frac{\hat{\sigma}^2}{\sum (x_i - \bar{x})^2}}
$$
$$
\sqrt{\mathbb{V}(\hat{\beta_0})} = \sqrt{\hat{\sigma}^2 \cdot \left( \frac{1}{n} + \frac{\bar{x}^2}{\sum (x_i - \bar{x})^2} \right)}
$$
where where $\hat{\sigma}^2 $ is the unbiased estimator for $\sigma^2 $ :
$$
\sigma^2 = \frac{\sum (y_i - \hat{y}_i)^2}{n - 2}
$$

***

#### (c). Determination Coefficient $R^2$
$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$$

```python
# Calculate means
x_mean = np.mean(log_gnp)
y_mean = np.mean(log_investment)

# Calculate beta1 (slope) and beta0 (intercept)
numerator = np.sum((log_gnp - x_mean) * (log_investment - y_mean))
denominator = np.sum((log_gnp - x_mean) ** 2)
beta1 = numerator / denominator
beta0 = y_mean - beta1 * x_mean

# Calculate predictions and residuals
y_pred = beta0 + beta1 * log_gnp
residuals = log_investment - y_pred

# Calculate residual variance (sigma^2)
n = len(log_gnp)
sigma_squared = np.sum(residuals**2) / (n - 2)

# Standard deviations of beta1 and beta0
std_beta1 = np.sqrt(sigma_squared / denominator)
std_beta0 = np.sqrt(
    sigma_squared * (1 / n + x_mean**2 / denominator)
)

# Determination coefficient (R^2)
ss_total = np.sum((log_investment - y_mean) ** 2)
ss_residual = np.sum(residuals ** 2)
r_squared = 1 - (ss_residual / ss_total)

# Output results
print("Linear Regression Results:")
print(f"Intercept (beta0): {beta0:.4f}")
print(f"Slope (beta1): {beta1:.4f}")
print(f"Standard deviation of intercept: {std_beta0:.4f}")
print(f"Standard deviation of slope: {std_beta1:.4f}")
print(f"Determination coefficient (R^2): {r_squared:.4f}")
```

### Question 4.

To test the significance of the slope ($\beta_1$) using the Student's $t$-test, we follow these steps:

***

#### (a) Hypotheses
- **Null Hypothesis ($H_0$)**: $\beta_1 = 0$ (No relationship between the predictor and response variables).
- **Alternative Hypothesis ($H_a$)**: $\beta_1 \neq 0$ (There is a relationship between the predictor and response variables).

***

#### (b) Test Statistic
The test statistic for the slope is calculated as:
$$
t = \frac{\hat{\beta_1}}{\sqrt{\mathbb{V}(\hat{\beta_1})}}
$$

***

#### (c) Confidence interval
For a given confidence level $\alpha$, the confidence interval for $t$:
$$
t \in \left[ - \mathcal{T}^{(n-2)}_{1-\alpha/2} , \;  + \mathcal{T}^{(n-2)}_{1-\alpha/2} \right]
$$
where $\mathcal{T}^{(n-2)}_{1-\alpha/2}$  critical value of the $t$-distribution with $n-2$ degrees of freedom, corresponding to the chosen confidence level.

***

#### (d) P-value
Using the $t$-statistic and degrees of freedom, the two-tailed p-value is computed as:
$$
\text{p-value} = 2 \cdot (1 - F(|t|, \text{df}))
$$
where $F$ is the cumulative distribution function of the $t$-distribution.

```python
# Calculate t-statistic for the slope
t_stat = beta1 / std_beta1

# Degrees of freedom
df = n - 2

# Two-tailed p-value
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

# Report results
print("Significance Test for the Slope:")
print(f"t-statistic: {t_stat:.4f}")
print(f"Degrees of freedom: {df}")
print(f"p-value: {p_value:.4e}")
```

### Question 5.

#### (a) Formula for Confidence Interval (CI):

The **Confidence Interval** for a predicted value at $x$ is given by:

$$
\text{CI}(x) = \hat{\beta}_0 + \hat{\beta}_1 x \pm \mathcal{T}^{(n-2)}_{1-\alpha/2} \cdot \sqrt{\hat{\sigma}^2 \left(\frac{1}{n} + \frac{(x - \bar{x})^2}{\sum_{i=1}^{n} (x_i - \bar{x})^2}\right)}
$$

***

#### (b) Formula for Prediction Interval (PI):

The **Prediction Interval** for a predicted value at $x$ is given by:

$$
\text{PI}(x) = \hat{\beta}_0 + \hat{\beta}_1 x \pm \mathcal{T}^{(n-2)}_{1-\alpha/2} \cdot \sqrt{\hat{\sigma}^2 \left( 1 + \frac{1}{n} + \frac{(x - \bar{x})^2}{\sum_{i=1}^{n} (x_i - \bar{x})^2}\right)}
$$

```python
gnp_value = 1000
log_gnp_value = np.log(gnp_value)

# Estimate the Investment for GNP = 1000
investment_estimate = beta0 + beta1 * log_gnp_value

# Calculate the critical t-value for a 90% confidence level
alpha = 0.10  # 90% confidence
t_critical = t.ppf(1 - alpha / 2, df)

# Calculate the sum of squared differences from the mean of X
sum_sq_diff = np.sum((log_gnp - x_mean) ** 2)

# Confidence Interval (CI) for the mean response
ci_lower = investment_estimate - t_critical * np.sqrt(sigma_squared * (1 / len(log_gnp) + (log_gnp_value - x_mean) ** 2 / sum_sq_diff))
ci_upper = investment_estimate + t_critical * np.sqrt(sigma_squared * (1 / len(log_gnp) + (log_gnp_value - x_mean) ** 2 / sum_sq_diff))

# Prediction Interval (PI) for a new observation
pi_lower = investment_estimate - t_critical * np.sqrt(sigma_squared * (1 + 1 / len(log_gnp) + (log_gnp_value - x_mean) ** 2 / sum_sq_diff))
pi_upper = investment_estimate + t_critical * np.sqrt(sigma_squared * (1 + 1 / len(log_gnp) + (log_gnp_value - x_mean) ** 2 / sum_sq_diff))

# Output the results
print(f"Estimated Investment for GNP = {gnp_value}: {np.exp(investment_estimate):.4f}")
print(f"90% Confidence Interval: [{np.exp(ci_lower):.4f}, {np.exp(ci_upper):.4f}]")
print(f"90% Prediction Interval: [{np.exp(pi_lower):.4f}, {np.exp(pi_upper):.4f}]")
```

### Question 9.

we need to compute the **Gram matrix** for the linear regression of Investment on GNP and Interest. The Gram matrix is computed as:

$$
X^\top X
$$

Where $ X $ is the Design Matrix includes :
   - A column of ones for the intercept.
   - A column for $ \log(\text{GNP}) $.
   - A column for $ \text{Interest} $ (not transformed).

   $$
   X = 
   \begin{bmatrix}
   1 & \log(\text{GNP}_1) & \text{Interest}_1 \\
   1 & \log(\text{GNP}_2) & \text{Interest}_2 \\
   \vdots & \vdots & \vdots \\
   1 & \log(\text{GNP}_n) & \text{Interest}_n
   \end{bmatrix}
   $$

Therefore :
   $$
   X^\top X = 
   \begin{bmatrix}
   1 & \log(\text{GNP}) & \text{Interest}
   \end{bmatrix}
   \cdot
   \begin{bmatrix}
   1 \\ \log(\text{GNP}) \\ \text{Interest}
   \end{bmatrix}
   $$

And in to check full rank we use the eigenvalues of $ X^\top X $ to check if all are positive.

```python
interest = data["interest"]

X = np.column_stack((np.ones(n), log_gnp, interest))

# Compute the Gram matrix
XtX = X.T @ X

# Check for full rank
eigenvalues = np.linalg.eigvals(XtX)
is_full_rank = np.all(eigenvalues > 0)

# Output results
print("Eigenvalues of Gram Matrix:", eigenvalues)
print("Is the Gram matrix of full rank?", "Yes" if is_full_rank else "No")
```

### Question 10. 

We consider the model:  
$$
Y = X \boldsymbol{\beta} + \boldsymbol{\varepsilon}, \quad \boldsymbol{\varepsilon} \sim \mathcal{N}(\boldsymbol{0}, \sigma^2 I_n),
$$  
In expanded form:
$$
\begin{bmatrix}
    y_1 \\
    y_2 \\
    \vdots \\
    y_n
\end{bmatrix}
=
\begin{bmatrix}
    1 & x_{11} & x_{12} & \cdots & x_{1(k-1)} \\
    1 & x_{21} & x_{22} & \cdots & x_{2(k-1)} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    1 & x_{n1} & x_{n2} & \cdots & x_{n(k-1)}
\end{bmatrix}
\begin{bmatrix}
    \beta_0 \\
    \beta_1 \\
    \beta_2 \\
    \vdots \\
    \beta_{k-1}
\end{bmatrix}
+
\begin{bmatrix}
    \varepsilon_1 \\
    \varepsilon_2 \\
    \vdots \\
    \varepsilon_n
\end{bmatrix}.
$$

---

####  (a) Estimation of Regression Coefficients $ \boldsymbol{\beta} $ :
The Ordinary Least Squares (OLS) solution minimizes the residual sum of squares $ \| Y - X\boldsymbol{\beta} \|_2^2 $. The estimated coefficients $ \hat{\boldsymbol{\beta}} $ are given by:  

$$
\hat{\boldsymbol{\beta}} = (X^\top X)^{-1} X^\top Y
$$  

---

#### (b) Residual Variance $ \hat{\sigma}^2 $ :  
The residual variance is calculated as:  

$$
\hat{\sigma}^2 = \frac{\boldsymbol{\varepsilon}^T \boldsymbol{\varepsilon}}{n - (p+1)} = \frac{1}{n - (p+1)} \|Y - \hat{Y} \|^2_2
$$  

where  $ \boldsymbol{\varepsilon} = Y - X\hat{\boldsymbol{\beta}} $ are the residuals.  

---

#### (c) Variance-Covariance Matrix of $ \hat{\boldsymbol{\beta}} $ :  
The variance-covariance matrix of the estimated coefficients $ \hat{\boldsymbol{\beta}} $ is given by:  

$$
\mathbb{V}(\hat{\boldsymbol{\beta}}) = \hat{\sigma}^2 (X^\top X)^{-1}
$$  

where diagonal entries of $ \mathbb{V}(\hat{\boldsymbol{\beta}}) $ correspond to $ \mathbb{V}(\hat{\beta}_j) $

---

####  (d) Confidence Interval for Predictions $ \text{CI}(x_*) $ : 
For a new observation $x_* $, the confidence interval for the predicted value is:  

$$
\text{CI}(x_*) = x_*^\top \hat{\boldsymbol{\beta}} \;\pm\; \mathcal{T}^{(n-p-1)}_{\alpha/2} \cdot \sqrt{\hat{\sigma}^2 \cdot x_*^T (X^T X)^{-1} x_*},
$$  

where:  
- $ x_* $ (a $ 1 \times (p+1) $ row vector, including the intercept).
- $ x_*^\top \hat{\boldsymbol{\beta}} $: Predicted value for $ x_* $.  
- $ \mathcal{T}^{(n-p-1)}_{\alpha/2} $: Critical value from the t-distribution with $ n-p-1 $ degrees of freedom.  
- $ \hat{\sigma}^2 x_*^T (X^T X)^{-1} x_* $: Variance of the prediction.

---

#### (e) Prediction Interval for Predictions $ \text{PI}(x_*) $ : 
The prediction interval accounts for both the variability in the predicted mean and the residual variability. It is calculated as: 

$$
\text{PI}(x_*) = x_*^T \hat{\boldsymbol{\beta}} \;\pm\; \mathcal{T}^{(n-p-1)}_{\alpha/2} \cdot \sqrt{\hat{\sigma}^2 (1 + x_*^T (X^T X)^{-1} x_*)}.
$$  

Here, the additional term $ 1 + x_*^T (X^\top X)^{-1} x_* $ reflects the uncertainty in predicting a new observation.

---

#### (f) Determination Coefficient $ R^2 $  
The coefficient of determination is defined as:  
$$
R^2 = 1 - \frac{\|Y - \hat{Y} \|^2_2}{\|Y - \bar{Y}\cdot 1 \|^2_2},
$$  

---

#### (g) Testing Significance of Coefficients ($ t $-test) :  
For each coefficient $ \beta_j $, the null hypothesis is $ H_0: \beta_j = 0 $. The test statistic is:  
$$
t_j = \frac{\hat{\beta}_j}{\sqrt{\mathbb{V}(\hat{\beta}_j})}
$$  
The p-value is computed as:  
$$
p_j = 2 \cdot (1 - F_t(|t_j|, \nu)),
$$  
where $ \nu = n - (p+1) $ is the degrees of freedom, and $ F_t $ is the cumulative distribution function of the t-distribution.

---


```python
p = X.shape[1]

Y = log_investment
XtX_inv = np.linalg.inv(XtX)

# Estimate Coefficients (beta)
beta = XtX_inv @ X.T @ Y

# Residuals and Determination Coefficient (R^2)
Y_pred = X @ beta
residuals = Y - Y_pred
SS_residuals = residuals.T @ residuals
SS_total = np.sum((Y - np.mean(Y))**2)
R_squared = 1 - (SS_residuals / SS_total)

# Variance of Residuals and Standard Errors
sigma_squared = SS_residuals / (n - p)
var_beta = sigma_squared * XtX_inv
std_errors = np.sqrt(np.diag(var_beta))

# t-Statistics and p-values
t_stats = beta.flatten() / std_errors
p_values = [2 * (1 - t.cdf(np.abs(ti), df=n - X.shape[1])) for ti in t_stats]

# Output Results
print("Regression Coefficients:")
for i, b in enumerate(beta.flatten()):
    print(f"Beta_{i}: {b:.4f}")

print("\nStandard Errors:")
for i, se in enumerate(std_errors):
    print(f"SE(Beta_{i}): {se:.4f}")

print("\nT-Statistics and P-Values:")
for i, (t_stat, p_val) in enumerate(zip(t_stats, p_values)):
    print(f"Beta_{i}: t = {t_stat:.4f}, p = {p_val}")

print(f"\nDetermination Coefficient (R^2): {R_squared:.4f}")
```

For $\beta_2$ The p-value is much greater than $\alpha = 0.05$, so the coefficient for Interest is not statistically significant. This suggests that Interest does not have a meaningful linear relationship with the Investment in this model.

### Question 11. 

Calculate and visualize the prediction of $\log(\text{Investment})$ for given values of $\log(\text{GNP})$ and Interest, along with 99.9% confidence and prediction intervals, using regression coefficients and residual variance.

```python
gnp_value = 1000
interest_value = 10

# Adding a new data point for prediction
new_data = np.array([[1, np.log(gnp_value), interest_value]])

# Prediction for log(Investment)
predicted_log_investment = new_data @ beta

# Confidence interval
alpha = 0.001  # 99.9% confidence level
t_critical = t.ppf(1 - alpha / 2, df=n - p)

# Confidence interval
variance_conf = sigma_squared * (new_data @ XtX_inv @ new_data.T)
std_error_conf = np.sqrt(variance_conf)

# Confidence interval bounds
ci_lower = predicted_log_investment - t_critical * std_error_conf
ci_upper = predicted_log_investment + t_critical * std_error_conf

# Prediction interval
std_error_pred = np.sqrt(sigma_squared + variance_conf)

# Prediction interval bounds
pi_lower = predicted_log_investment - t_critical * std_error_pred
pi_upper = predicted_log_investment + t_critical * std_error_pred

# Print results
print(f"Predicted Investment for GNP = {gnp_value} and Interest = {interest_value}: {np.exp(predicted_log_investment[0]):.4f}")
print(f"99.9% Confidence Interval: [{np.exp(ci_lower[0, 0]):.4f}, {np.exp(ci_upper[0, 0]):.4f}]")
print(f"99.9% Prediction Interval: [{np.exp(pi_lower[0, 0]):.4f}, {np.exp(pi_upper[0, 0]):.4f}]")
```

### Question 12. 

The code generates a 3D plot displaying the regression plane, original data points, and 99.9% confidence interval surfaces for log-transformed GNP, interest rates, and log-transformed investment values. 

```python
gnp = data["gnp"]
investment = data["invest"]
interest = data["interest"]

log_gnp = np.log(gnp)
log_investment = np.log(investment)
```

```python
# Generate a grid of log(GNP) and Interest values for the plane and CI surfaces
log_gnp_range = np.linspace(log_gnp.min(), log_gnp.max(), 50)
interest_range = np.linspace(interest.min(), interest.max(), 50)
log_gnp_grid, interest_grid = np.meshgrid(log_gnp_range, interest_range)

# Predictions for the regression plane
X_grid = np.column_stack((
    np.ones(log_gnp_grid.size),
    log_gnp_grid.ravel(),
    interest_grid.ravel()
))
log_investment_pred = X_grid @ beta

# Variance of predictions
var_pred = np.array([
    x @ XtX_inv @ x.T for x in X_grid
]).reshape(log_gnp_grid.shape) * sigma_squared

# Confidence intervals (99.9%)
t_critical = t.ppf(0.9995, df=n - p)
ci_upper = log_investment_pred.reshape(log_gnp_grid.shape) + t_critical * np.sqrt(var_pred)
ci_lower = log_investment_pred.reshape(log_gnp_grid.shape) - t_critical * np.sqrt(var_pred)

# Plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of original data points
ax.scatter(log_gnp, interest, log_investment, color='blue', label='Data Points')

# Regression plane
ax.plot_surface(
    log_gnp_grid, interest_grid, log_investment_pred.reshape(log_gnp_grid.shape),
    color='red', alpha=0.5, label='Regression Plane'
)

# Confidence interval surfaces
ax.plot_surface(
    log_gnp_grid, interest_grid, ci_upper, color='green', alpha=0.3, label='99.9% CI Upper Surface'
)
ax.plot_surface(
    log_gnp_grid, interest_grid, ci_lower, color='green', alpha=0.3, label='99.9% CI Lower Surface'
)

# Labels and legend
ax.set_xlabel('Log(GNP)')
ax.set_ylabel('Interest')
ax.set_zlabel('Log(Investment)')
plt.title('3D Regression Plane and 99.9% Confidence Interval Surfaces')
plt.legend(loc='best')
plt.show()
```
